```python
#SOLUTION
# Create new column to check age groups

def age_groups(series):
    if series <20:
        return "0-19 years"
    elif 20 <= series < 30:
        return "20-29"
    elif 30 <= series < 40:
        return "30-39 years"
    elif 40 <= series < 50:
        return "40-49"
    elif 50 <= series < 60:
        return "50-59"
    else:
        return "over 60"
    
data["age_group"] = data["age"].apply(age_groups)  #SOLUTION
```
